```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  
```

### GA

```python
cp = pd.DataFrame()
opt = 'GA'
for p in [100]:
    for ma in [50, 20]:
        for mu in [10, 5]:
            cpfile = 'ccr/cc-%s-99--%d-%d-%d-mse.csv' % (opt, p, ma, mu)
            tmp = pd.read_csv(cpfile, names=["i", "train", "test", "valid"])
            tmp["p"] = p
            tmp["ma"] = ma
            tmp["mu"] = mu
            tmp["opt"] = opt
            tmp["params"] = "pop=%d, mate=%d, mutate=%d" % (p, ma, mu)
            if cp.shape[0] == 0:
                cp = tmp.copy(deep=True)
            else:
                cp = cp.append(tmp)
```

```python
cp[cp["train"] == cp["train"].min()]
```

```python
for tt in ['train', 'test']:
    legends = []
    name = cp["opt"].unique().item()
    for p in cp["params"].unique().tolist():
        df = cp[cp["params"] == p]
        df = df[df["i"].mod(25) == 0]
        legends.append(p)
        plt.plot(df["i"], df[tt], '.-')
    plt.legend(legends, loc="best")
    plt.xlabel("iteration")
    plt.ylabel("mse")
    plt.title(name+" "+tt+" performance")
    plt.gcf()
    plt.savefig('cc-'+name+'-'+tt+'-mse.png')
    plt.show()
```

```python
opt = 'GA'
p=100
ma=50
mu=5
cpfile = 'ccr/cc-%s-99--%d-%d-%d-mse.csv' % (opt, p, ma, mu)
ga = pd.read_csv(cpfile, names=["i", "train", "test", "valid"])
#ga = ga[ga["i"].mod(25) == 0]
ga["p"] = p
ga["ma"] = ma
ga["mu"] = mu
ga["params"] = "pop=%d, mate=%d, mutate=%d" % (p, ma, mu)
ga["opt"] = opt
```

### SA

```python
cp = pd.DataFrame()
opt = 'SA'
for t in ["1E10", "100."]:
    #for ma in [100, 50, 10]:
        for c in ["0.25", "0.5", "0.75", "0.95"]:
            cpfile = 'ccr/cc-%s-99--%s-%s-mse.csv' % (opt, t, c)
            tmp = pd.read_csv(cpfile, names=["i", "train", "test", "valid"])
            tmp["t"] = t
            tmp["cooling"] = c
            tmp["params"] = "t=%s, cooling=%s" % (t, c)
            tmp["opt"] = opt
            if cp.shape[0] == 0:
                cp = tmp.copy(deep=True)
            else:
                cp = cp.append(tmp)
```

```python
cp[cp["train"] == cp["train"].min()]
```

```python
for tt in ['train', 'test']:
    legends = []
    name = cp["opt"].unique().item()
    for p in cp["params"].unique().tolist():
        df = cp[cp["params"] == p]
        df = df[df["i"].mod(50) == 0]
        legends.append(p)
        plt.plot(df["i"], df[tt], '.-')
    plt.legend(legends, loc="best")
    plt.xlabel("iteration")
    plt.ylabel("mse")
    plt.title(name+" "+tt+" performance")
    plt.gcf()
    plt.savefig('cc-'+name+'-'+tt+'-mse.png')
    plt.show()
```

```python
opt = 'SA'
t="100."
c="0.75"
cpfile = 'ccr/cc-%s-99--%s-%s-mse.csv' % (opt, t, c)
sa = pd.read_csv(cpfile, names=["i", "train", "test", "valid"])
#sa = sa[sa["i"].mod(50) == 0]
sa["t"] = t
sa["cooling"] = c
sa["params"] = "t=%s, cooling=%s" % (t, c)
sa["opt"] = opt
```

```python
legend = []
loop = 0
for i,v in enumerate(['train', 'test']):
    for out in [ga, bp, rhc, sa]:
        out = out[out["i"].mod(50) == 0]
        out = out[out["i"] < 500]
        x = out.iloc[:,0]
        y = out.iloc[:,i+1]
        plt.plot(x, y, '.-')
        legend.append(out.iloc[0,-1])
        loop += 1

    plt.legend(legend, loc="best")
    plt.ylabel('mse')
    plt.xlabel('iteration')
    plt.title('best param performance on '+v+' set')
    plt.gcf()
    plt.savefig('cc'+v+'mse.png')
    plt.show()
```
